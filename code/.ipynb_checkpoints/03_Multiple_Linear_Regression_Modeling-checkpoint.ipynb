# Multiple Linear Regression Modeling

Multiple Linear Regression Modeling models the relationship between explanatory variables and the variable which we wish to predict. It does this by assigning each explanatory variables a weighting coefficient called a beta coefficient, and multiplies each beta coefficient by each explanatory variable value and then adding the residual in order to calculate a prediction. 

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pickle
import numpy as np
np.random.seed(42)

%matplotlib inline
```

This loads in the libraries we need. 

```python
X_train_sc = np.load('../datasets/X_train_sc.npy')
X_test_sc = np.load('../datasets/X_test_sc.npy')
y_train = np.load('../datasets/y_train.npy')
y_test = np.load('../datasets/y_test.npy')
```

This imports the scaled test and training data we generated earlier. 

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_sc, y_train)
```

```python
lr.score(X_train_sc, y_train)
```

```python
lr.score(X_test_sc, y_test)
```

The extreme lowness of my scaled test data is concerning. 

It appears that I can predict 89 percent of the training set variance using my lasso regression model. 

```python
def RMSE(true, predicted):
    diff = true - predicted
    squared_diff = np.square(diff)
    return np.mean(squared_diff) ** 0.5
```

```python
RMSE(y_test,lr.predict(X_test_sc))
```

The largeness of my RMSE means that I am actually doing worse here than with the naive model. 

```python
columns = np.load('../datasets/columns.npy')
```

```python
betas = pd.DataFrame(lr.coef_)
```

```python
betas.set_index(columns,inplace=True)
```

```python
betas.columns = ['weights']
```

```python
betas['abs_w'] = betas.weights.abs()
```

```python
beta_0 = lr.intercept_
```

```python
weights = betas.sort_values('abs_w', ascending=False)['weights'].head(8)
```

```python
labels = weights.index
```

```python
plt.figure(figsize=(20,10))
plt.barh(labels, weights)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel('Weights', fontsize=30)
plt.title(f'Most Important Features and Weights\ny intercept: ${beta_0:0.02f}', fontsize=42)
plt.tight_layout()
```

```python
def md_coef_table(labels, weights):
    print('| feature | weight |\n| --- | --- |')
    for i in range(len(labels)):
        print(f'| {labels[i]} | {weights[i]:.2e} |')
```

```python
md_coef_table(labels, weights)
```

From my preliminary analysis, I surmise that the magnitude of the weights, being on a scale of 10 to the sixteenth power, means that my baseline model is not sufficient for accurate prediction of my results. Therefore, there would be no use in talking about which features are the most and least helpful towards predicting home sale price.
