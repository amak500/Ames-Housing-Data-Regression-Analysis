# Ridge Regression Modeling

Rigdge Regularization, or L2 Regression, is a form of linear regression that seeks to reduce multicollinearity, or nonindependence or predictor variables. It does this by making the sum of the squares of the differences between the predicted and actual values as small as possible. 

```python
import pandas as pd
from sklearn.linear_model import Ridge, RidgeCV
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import pickle
import numpy as np
np.random.seed(42)

%matplotlib inline
```

This imports the libraries we need to do ridge regression and read in our files.

```python
X = pd.read_pickle("../datasets/training_data_cleaned_X.pkl")
y = pd.read_pickle("../datasets/training_data_cleaned_y.pkl")
X_train_sc = np.load('../datasets/X_train_sc.npy')
X_test_sc = np.load('../datasets/X_test_sc.npy')
y_train = np.load('../datasets/y_train.npy')
y_test = np.load('../datasets/y_test.npy')
```

This reads in our files. 

```python
# Set up a list of ridge alphas to check.
r_alphas = np.logspace(0, 5, 100)
# Generates 100 values equally between 0 and 5,
# then converts them to alphas between 10^0 and 10^5.

# Cross-validate over our list of ridge alphas.
ridge_model = RidgeCV(alphas=r_alphas, scoring='r2', cv=5)

# Fit model using best ridge alpha!
ridge_model = ridge_model.fit(X_train_sc, y_train)
```

```python
ridge_optimal_alpha = ridge_model.alpha_
ridge_optimal_alpha
```

This prints out the optimal alpha value, i.e. 475.0810162102798.

```python
ridge_model_preds = ridge_model.predict(X_test_sc)
ridge_model_preds_train = ridge_model.predict(X_train_sc)
```

```python
ridge_model_preds_r2 = r2_score(y_test, ridge_model_preds)
ridge_model_preds_train_r2 = r2_score(y_train, ridge_model_preds_train)
print(f"The r2 score of the training set is {ridge_model_preds_train_r2}.")
print(f"The r2 score of the test set is {ridge_model_preds_r2}.")
```

It appears that I can predict 87 percent of the training set variance and 90 percent of the test set variance using my ridge regression model. The fact that my test score r2 is higher than my training score r2 might indicated slight underfitting on the part of my model. 

```python
def RMSE(true, predicted):
    diff = true - predicted
    squared_diff = np.square(diff)
    return np.mean(squared_diff) ** 0.5
```

```python
RMSE(y_test,ridge_model.predict(X_test_sc))
```

Additionally, my RMSE is roughly 24601.381265834087, the lowest of the four models I've used. Thus, I will choose to use it as my production model.  

```python
columns = np.load('../datasets/columns.npy')
```

```python
betas = pd.DataFrame(ridge_model.coef_)
```

```python
betas.set_index(columns,inplace=True)
```

```python
betas.columns = ['weights']
```

```python
betas['abs_w'] = betas.weights.abs()
```

```python
betas['abs_w'].sort_values(ascending=False).head(10)
```

```python
beta_0 = ridge_model.intercept_
```

```python
weights = betas.sort_values('abs_w', ascending=False)['weights'].head(8)
```

```python
labels = weights.index
```

```python
plt.figure(figsize=(20,10))
plt.barh(labels, weights)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel('Weights', fontsize=30)
plt.title(f'Most Important Features and Weights\ny intercept: ${beta_0:0.02f}', fontsize=42)
plt.tight_layout()
```

This creates a graph of the features that are most important towards determining sale price and their associated weights.  

```python
reverse_weights = betas.sort_values('abs_w', ascending=True)['weights'].head(8)
reverse_labels = reverse_weights.index
plt.figure(figsize=(20,10))
plt.barh(reverse_labels, reverse_weights)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel('Weights', fontsize=30)
plt.title(f'Least Important Features and Weights\ny intercept: ${beta_0:0.02f}', fontsize=42)
plt.tight_layout()
```

```python
def md_coef_table(labels, weights):
    print('| feature | weight |\n| --- | --- |')
    for i in range(len(labels)):
        print(f'| {labels[i]} | {weights[i]:.2e} |')
```

This is a function designed to created a markdown table. 

```python
md_coef_table(labels, weights)
```

The magnitude of my weights being on the scale of 10 to the third power seems reasonable. Additionally, the above graphs and diagrams show that the most important features according to this model for predicting sale prices are all related to the quality of a house, be it overall, or of various features such as first floor square footage or the kitchen. The least important features seemed to be tied to when the property was sold and the property's exterior.
